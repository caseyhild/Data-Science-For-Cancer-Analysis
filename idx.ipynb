```python
data_cancer = pd.read_csv('data/cancer_reg.csv', index_col=False)
new_columns = data_cancer.columns.str.lower()
new_data_cancer = data_cancer.copy()
new_data_cancer.columns = new_columns
new_data_cancer = new_data_cancer[['geography', 'avganncount', 'avgdeathsperyear', 'popest2015', 'medincome', 'binnedinc', 'medianage', 'medianagemale', 'medianagefemale', 'pcths25_over', 'pctbachdeg25_over', 'pctemployed16_over','pctunemployed16_over', 'pctprivatecoverage', 'pctpubliccoverage', 'povertypercent', 'studypercap']]
new_data_cancer[['county', 'state']] = new_data_cancer['geography'].str.split(',', n=1, expand=True)
new_data_cancer = new_data_cancer.drop(['geography'], axis=1)
county_col = new_data_cancer.pop('county')
state_col = new_data_cancer.pop('state')
new_data_cancer.insert(0, 'county', county_col)
new_data_cancer.insert(1, 'state', state_col)
new_data_cancer.head()

new_data_cancer.to_csv('data/cancer_reg_clean.csv', index=False)
```
